```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
from pandas._config import dates
import pandas as pd
```

# EJERCICIO 1 - Análisis de Tiempo de Respuesta de Servidores Web

### Contexto
Un equipo de analistas de datos en una empresa de tecnología ha sido asignado para evaluar el rendimiento de dos servidores web. Los tiempos de respuesta, medidos en milisegundos, son críticos para la experiencia del usuario y el negocio en línea.


```python
Servidor_1 = [200, 220, 210, 250, 275, 230, 240, 255, 280, 220, 210, 195]
Servidor_2= [180, 190, 175, 193, 200, 205, 201, 210, 195, 185, 176, 199]
```

### 1. Utilizar Python para calcular y graficar un histograma y un gráfico de densidad de los tiempos de respuesta de ambos servidores. Superponer la media y la mediana en los gráficos.

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

sns.histplot(Servidor_1, kde=True, ax=axs[0])
axs[0].axvline(np.mean(Servidor_1), color='r', linestyle='--')
axs[0].axvline(np.median(Servidor_1), color='g', linestyle='-')
axs[0].set_title(f'Servidor 1 (Media: {np.mean(Servidor_1):.2f}, Mediana: {np.median(Servidor_1)})')

sns.histplot(Servidor_2, kde=True, ax=axs[1])
axs[1].axvline(np.mean(Servidor_2), color='r', linestyle='--')
axs[1].axvline(np.median(Servidor_2), color='g', linestyle='-')
axs[1].set_title(f'Servidor 2 (Media: {np.mean(Servidor_2):.2f}, Mediana: {np.median(Servidor_2)})')

plt.tight_layout()
plt.show()
```

```python
media_servidor1 = np.mean(Servidor_1)
mediana_servidor1 = np.median(Servidor_1)

media_servidor2 = np.mean(Servidor_2)
mediana_servidor2 = np.median(Servidor_2)

plt.subplot(1, 2, 1)
sns.kdeplot(Servidor_1, fill=True, color='orange')
plt.axvline(media_servidor1, color='r', linestyle='--', label='Media')
plt.axvline(mediana_servidor1, color='g', linestyle='-', label='Mediana')
plt.title('Gráfico de Densidad - Servidor 1')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.legend()

plt.subplot(1, 2, 2)
sns.kdeplot(Servidor_2, fill=True, color='blue')
plt.axvline(media_servidor2, color='r', linestyle='--', label='Media')
plt.axvline(mediana_servidor2, color='g', linestyle='-', label='Mediana')
plt.title('Gráfico de Densidad - Servidor 2')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.legend()

plt.tight_layout()
plt.show()
```

### 2. Calcular el sesgo de cada conjunto de datos y discutir su implicación en la estrategia de optimización del servidor.

```python
sesgo_S1 = skew(Servidor_1)
sesgo_S2 = skew(Servidor_2)

print(sesgo_S1)
print(sesgo_S2)
```

Al considerar el sesgo de los datos, se puede concluir que el primer servidor, claramente sesgado hacia el lado positivo, como se evidencia por la dirección de la cola en el gráfico, tiende a tener tiempos de respuesta considerablemente mayores en comparación con el segundo servidor, que muestra una tendencia hacia tiempos de respuesta más bajos.

La diferencia entre los tiempos de respuesta promedio de ambos servidores es significativa, lo que sugiere que el Servidor_2 es más consistente en mantener tiempos de respuesta más bajos.

### 3. Calcular y comparar la varianza y la desviación estándar de los tiempos de respuesta en ambos servidores.

```python
varianzaA = np.var(Servidor_1)
desviacionEstandarA = np.std(Servidor_1)

varianzaB = np.var(Servidor_2)
desviacionEstandarB = np.std(Servidor_2)

print("Servidor 1:")
print("Varianza:", varianzaA)
print("Desviación Estándar:", desviacionEstandarA)

print("\nServidor 2:")
print("Varianza:", varianzaB)
print("Desviación Estándar:", desviacionEstandarB)
```

### 4. Suponiendo que el objetivo de la empresa es tener un tiempo de respuesta medio menor a 200 ms, ¿qué servidor es más adecuado para cumplir con este objetivo? Justifique su respuesta

El tiempo de respuesta promedio del Servidor_2 es de 194.0 ms, lo que se encuentra por debajo del umbral objetivo de 200 ms cumpliendo asi con el objetivo de mantener una media se respuesta menor a 200 ms.

En contraste, el Servidor_1 tiene un tiempo de respuesta promedio superior, registrando 235.83 ms, lo que indica que su rendimiento medio excede el límite deseado.

# EJERCICIO 2 - Análisis de Tasas de Clics en Campañas Publicitarias

### Contexto
El equipo de marketing de una empresa de comercio electrónico está realizando A/B testing en dos versiones diferentes de un anuncio. La métrica clave de interés es la tasa de clics (CTR, por sus siglas en inglés).


```python
datosAnuncioA =  [0.1, 0.12, 0.11, 0.09, 0.1, 0.13, 0.12, 0.1, 0.14]
datosAnuncioB = [0.2, 0.18, 0.19, 0.21, 0.17, 0.16, 0.2, 0.22]
```

### 1. Utilizar Python para calcular y graficar un histograma y un gráfico de densidad de las tasas de clics de ambos anuncios. Superponer la media y la mediana en los gráficos.


```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

sns.histplot(datosAnuncioA, kde=True, ax=axs[0])
axs[0].axvline(np.mean(datosAnuncioA), color='r', linestyle='--')
axs[0].axvline(np.median(datosAnuncioA), color='g', linestyle='-')
axs[0].set_title(f'Anuncio A (Media: {np.mean(datosAnuncioA):.2f}, Mediana: {np.median(datosAnuncioA)})')

sns.histplot(datosAnuncioB, kde=True, ax=axs[1])
axs[1].axvline(np.mean(datosAnuncioB), color='r', linestyle='--')
axs[1].axvline(np.median(datosAnuncioB), color='g', linestyle='-')
axs[1].set_title(f'Anuncio B (Media: {np.mean(datosAnuncioB):.2f}, Mediana: {np.median(datosAnuncioB)})')

plt.tight_layout()
plt.show()
```

```python
mediaAnuncioA = np.mean(datosAnuncioA)
medianaAnuncioA = np.median(datosAnuncioA)

mediaAnuncioB = np.mean(datosAnuncioB)
medianaAnuncioB = np.median(datosAnuncioB)

plt.subplot(1, 2, 1)
sns.kdeplot(datosAnuncioA, fill=True, color='orange')
plt.axvline(mediaAnuncioA, color='r', linestyle='--', label='Media')
plt.axvline(medianaAnuncioA, color='g', linestyle='-', label='Mediana')
plt.title('Gráfico de Densidad - Anuncio A')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.legend()

plt.subplot(1, 2, 2)
sns.kdeplot(datosAnuncioB, fill=True, color='blue')
plt.axvline(mediaAnuncioB, color='r', linestyle='--', label='Media')
plt.axvline(medianaAnuncioB, color='g', linestyle='-', label='Mediana')
plt.title('Gráfico de Densidad - Anuncio B')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.legend()

plt.tight_layout()
plt.show()
```

### 2. Calcular el sesgo de cada conjunto de datos y discutir cómo esto podría influir en la estrategia publicitaria de la empresa.

```python
sesgoAnuncioA = skew(datosAnuncioA)
sesgoAnuncioB = skew(datosAnuncioB)

print(sesgoAnuncioA)
print(sesgoAnuncioB)
```

El sesgo en el anuncioA es positivo ya que la cola va hacia la derecha, lo cual nos puede decir que las personas dan mas click en ese anuncio, lo cual es un oportunidad para que implementen mejor marcketing en el anuncioA.

# EJERCICIO 4 - Análisis de Duración de Sesiones de Usuarios en una Plataforma de Streaming

### Contexto
Una popular plataforma de streaming desea entender cómo sus usuarios interactúan con su contenido. Específicamente, quieren saber cuánto tiempo pasan los usuarios en la plataforma durante una sesión.


```python
# Duración de sesiones en minutos
Género_Comedia = [35, 40, 45, 37, 39, 38, 42, 43, 50]
Género_Acción = [60, 65, 70, 58, 61, 63, 67]
```

### 1. Utilizar Python para calcular y graficar un histograma y un gráfico de densidad de la duración de sesiones de ambos géneros. Superponer la media y la mediana en los gráficos.

```python
 # HISTOGRAMA
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

sns.histplot(Género_Comedia, kde=True, ax=axs[0])
axs[0].axvline(np.mean(Género_Comedia), color='r', linestyle='--')
axs[0].axvline(np.median(Género_Comedia), color='g', linestyle='-')
axs[0].set_title(f'Género Comedia (Media: {np.mean(Género_Comedia):.2f}, Mediana: {np.median(Género_Comedia)})')

sns.histplot(Género_Acción, kde=True, ax=axs[1])
axs[1].axvline(np.mean(Género_Acción), color='r', linestyle='--')
axs[1].axvline(np.median(Género_Acción), color='g', linestyle='-')
axs[1].set_title(f'Género Acción (Media: {np.mean(Género_Acción):.2f}, Mediana: {np.median(Género_Acción)})')

plt.tight_layout()
plt.show()
```

```python
media_Género_Comedia = np.mean(Género_Comedia)
mediana_Género_Comedia = np.median(Género_Comedia)

media_Género_Acción = np.mean(Género_Acción)
mediana_Género_Acción = np.median(Género_Acción)

plt.subplot(1, 2, 1)
sns.kdeplot(Género_Comedia, fill=True, color='orange')
plt.axvline(media_Género_Comedia, color='r', linestyle='--', label='Media')
plt.axvline(mediana_Género_Comedia, color='g', linestyle='-', label='Mediana')
plt.title('Gráfico de Densidad - Género Comedia')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.legend()

plt.subplot(1, 2, 2)
sns.kdeplot(Género_Acción, fill=True, color='blue')
plt.axvline(media_Género_Acción, color='r', linestyle='--', label='Media')
plt.axvline(mediana_Género_Acción, color='g', linestyle='-', label='Mediana')
plt.title('Gráfico de Densidad - Género Acción')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.legend()

plt.tight_layout()
plt.show()
```

### 2. Calcular el sesgo de cada conjunto de datos y discutir su relevancia para la estrategia de contenido de la plataforma.

```python
sesgo_Género_Comedia = skew(Género_Comedia)
sesgo_Género_Acción = skew(Género_Acción)

print("Sesgo de Género Comedia:", sesgo_Género_Comedia)
print("Sesgo de Género Acción:", sesgo_Género_Acción)
```

Al analizar las gráficas, se hace evidente un claro sesgo positivo en ambos géneros, y sus valores medios y medianos están prácticamente superpuestos. Teniendo en cuenta esta información, podemos concluir que ambos géneros son bien recibidos.Viendo las graficas podemos ver un rango mas alto en el genero de accion eso sugiere un sesgo hacia sesiones mas largas en el genero de accion.
