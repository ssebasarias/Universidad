```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Grupo 1: Presión arterial normal
media1, desviacion1 = 120, 10
grupo1 = np.random.normal(media1, desviacion1, 1000)

# Grupo 2: Presión arterial alta
media2, desviacion2 = 140, 12
grupo2 = np.random.normal(media2, desviacion2, 1000)

# Calculamos el sesgo para cada grupo
sesgo1 = skew(grupo1)
sesgo2 = skew(grupo2)

print(f"Sesgo del primer grupo es: {sesgo1}")
print(f"Sesgo del segundo grupo es: {sesgo2}")
```

```python
# Crear subplots
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Grupo 1
sns.histplot(grupo1, kde=True, ax=axs[0])
axs[0].axvline(np.mean(grupo1), color='r', linestyle='--')
axs[0].axvline(np.median(grupo1), color='g', linestyle='-')
axs[0].set_title(f'Grupo 1 (Sesgo: {sesgo1:.2f})')

# Grupo 2
sns.histplot(grupo2, kde=True, ax=axs[1])
axs[1].axvline(np.mean(grupo2), color='r', linestyle='--')
axs[1].axvline(np.median(grupo2), color='g', linestyle='-')
axs[1].set_title(f'Grupo 2 (Sesgo: {sesgo2:.2f})')

plt.show()
```

# **Parte 1: Análisis Descriptivo**

```python
import numpy as np

# Datos proporcionados
tiempo_en_sitio_web = np.array([2, 3, 4, 5, 3, 4, 3, 2, 5, 10, 11, 12, 10, 5, 2, 1, 8, 10])
gasto_del_cliente = np.array([100, 101, 150, 200, 250, 200, 180, 210, 275, 280, 300, 325, 330])

# Cálculo de la media
media_tiempo = np.mean(tiempo_en_sitio_web)
media_gasto = np.mean(gasto_del_cliente)

# Cálculo de la mediana
mediana_tiempo = np.median(tiempo_en_sitio_web)
mediana_gasto = np.median(gasto_del_cliente)

# Cálculo de la varianza
varianza_tiempo = np.var(tiempo_en_sitio_web)
varianza_gasto = np.var(gasto_del_cliente)

# Cálculo de la desviación estándar
desviacion_tiempo = np.std(tiempo_en_sitio_web)
desviacion_gasto = np.std(gasto_del_cliente)

# Cálculo del CV

Cv_tiempo=desviacion_tiempo*100/media_tiempo
# Mostrar resultados
print(f"Tiempo en el sitio web: Media = {media_tiempo}, Mediana = {mediana_tiempo}, Varianza = {varianza_tiempo}, Desviación estándar = {desviacion_tiempo}, CV={Cv_tiempo}")
print(f"Gasto del cliente: Media = {media_gasto}, Mediana = {mediana_gasto}, Varianza = {varianza_gasto}, Desviación estándar = {desviacion_gasto}")
```

# **Interpretación**

**Tiempo en el sitio web:**

* La media de aproximadamente 5.5 minutos y la mediana de 4.5 minutos sugieren que la mayoría de los clientes pasan un tiempo moderado en el sitio web.
* La varianza y la desviación estándar altas indican que hay una variabilidad considerable en cuánto tiempo pasan los clientes en el sitio.

**Gasto del cliente:**

* La media de aproximadamente $223$ y la mediana de $210$ sugieren que los clientes suelen gastar alrededor de $210-$223.
* La varianza y la desviación estándar también son relativamente altas, lo que indica que hay clientes que gastan significativamente más que otros.

En general, estos datos podrían sugerir que aunque el tiempo medio en el sitio es moderado, hay un grupo de clientes que pasa mucho más tiempo y probablemente también gasta más. Estos podrían ser los clientes a los que la empresa querría dirigir ofertas o programas de lealtad.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
tiempo_en_sitio_web = np.array([2, 3, 4, 5, 3, 4, 3, 2, 5, 10, 11, 12, 10, 5, 2, 1, 8, 10])
gasto_del_cliente = np.array([100, 101, 150, 200, 250, 200, 180, 210, 275, 280, 300, 325, 330])

# Cálculo de estadísticas descriptivas
media_tiempo = np.mean(tiempo_en_sitio_web)
mediana_tiempo = np.median(tiempo_en_sitio_web)

media_gasto = np.mean(gasto_del_cliente)
mediana_gasto = np.median(gasto_del_cliente)

# Crear subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Histograma para tiempo en el sitio web
sns.histplot(tiempo_en_sitio_web, kde=False, ax=axs[0, 0])
axs[0, 0].axvline(media_tiempo, color='r', linestyle='--')
axs[0, 0].axvline(mediana_tiempo, color='g', linestyle='-')
axs[0, 0].set_title('Histograma del Tiempo en el Sitio Web')

# Gráfico de densidad para tiempo en el sitio web
sns.kdeplot(tiempo_en_sitio_web, fill=True, ax=axs[0, 1])
axs[0, 1].axvline(media_tiempo, color='r', linestyle='--')
axs[0, 1].axvline(mediana_tiempo, color='g', linestyle='-')
axs[0, 1].set_title('Gráfico de Densidad del Tiempo en el Sitio Web')

# Histograma para gasto del cliente
sns.histplot(gasto_del_cliente, kde=False, ax=axs[1, 0])
axs[1, 0].axvline(media_gasto, color='r', linestyle='--')
axs[1, 0].axvline(mediana_gasto, color='g', linestyle='-')
axs[1, 0].set_title('Histograma del Gasto del Cliente')

# Gráfico de densidad para gasto del cliente
sns.kdeplot(gasto_del_cliente, fill=True, ax=axs[1, 1])
axs[1, 1].axvline(media_gasto, color='r', linestyle='--')
axs[1, 1].axvline(mediana_gasto, color='g', linestyle='-')
axs[1, 1].set_title('Gráfico de Densidad del Gasto del Cliente')

plt.tight_layout()
plt.show()
```

```python
# Cálculo del sesgo
sesgo_tiempo = skew(tiempo_en_sitio_web)
sesgo_gasto = skew(gasto_del_cliente)

# Cálculo de la varianza y la desviación estándar
varianza_tiempo = np.var(tiempo_en_sitio_web)
varianza_gasto = np.var(gasto_del_cliente)

desviacion_tiempo = np.std(tiempo_en_sitio_web)
desviacion_gasto = np.std(gasto_del_cliente)

print(f"Sesgo del tiempo en el sitio web: {sesgo_tiempo:.2f}")
print(f"Sesgo del gasto del cliente: {sesgo_gasto:.2f}")

print(f"Varianza del tiempo en el sitio web: {varianza_tiempo:.2f}")
print(f"Varianza del gasto del cliente: {varianza_gasto:.2f}")

print(f"Desviación estándar del tiempo en el sitio web: {desviacion_tiempo:.2f}")
print(f"Desviación estándar del gasto del cliente: {desviacion_gasto:.2f}")
```

**Interpretación**

**Sesgo:**

1. Tiempo en el sitio web: El sesgo es positivo, lo que sugiere que hay una cola hacia la derecha. Esto podría indicar que algunos clientes pasan más tiempo del habitual en el sitio web, lo que podría ser una oportunidad para marketing dirigido.
2. Gasto del cliente: El sesgo es negativo, lo que indica una cola hacia la izquierda. Esto sugiere que la mayoría de los clientes gastan una cantidad considerable, pero hay unos pocos que gastan menos.

**Varianza y Desviación Estándar:**

3. La varianza y la desviación estándar del gasto del cliente son más altas en comparación con el tiempo pasado en el sitio web, lo que indica que los datos del gasto están más dispersos.
