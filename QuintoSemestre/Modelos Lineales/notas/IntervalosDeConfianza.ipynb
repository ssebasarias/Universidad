# **FÃ³rmulas**

* **Error EstÃ¡ndar (SE):** El error estÃ¡ndar es una medida de cuÃ¡nto se espera que varÃ­e la media de la muestra en torno a la media de la poblaciÃ³n. Se calcula como:

<center>$\text{SE} = \frac{\sigma}{\sqrt{n}}$</center>

* **Margen de Error (E):** El margen de error se calcula multiplicando el valor Z correspondiente al nivel de confianza deseado ($Z_{\alpha/2}$) por el error estÃ¡ndar:


<center>â$\text{E} = Z_{\alpha/2} \times \text{SE}$</center>

<center>$\text{LÃ­mite Inferior} = \bar{x} - E$</center>

<center>$\text{LÃ­mite Superior} = \bar{x} + E$</center>

# **Ejercicio 1: Tiempo de Carga de PÃ¡gina Web**
Una empresa de tecnologÃ­a que ofrece soluciones de hosting quiere medir el tiempo de carga de las pÃ¡ginas web de sus clientes. DespuÃ©s de medir el tiempo de carga de una muestra de 50 pÃ¡ginas web, la empresa encuentra que el tiempo medio de carga es de 2.3 segundos. La desviaciÃ³n estÃ¡ndar conocida del tiempo de carga es de 0.5 segundos.

Calcule un intervalo de confianza del 95% para el tiempo medio de carga de todas las pÃ¡ginas web alojadas por la empresa.

```python
from scipy.stats import norm

def calcular_intervalo_confianza(media_muestral, desviacion_poblacional, tamano_muestra, nivel_confianza):
    # Calcular el valor crÃ­tico de Z usando el nivel de confianza
    valor_critico = norm.ppf(1 - (1 - nivel_confianza) / 2)

    # Calcular el error estÃ¡ndar
    error_estandar = desviacion_poblacional / (tamano_muestra ** 0.5)

    # Calcular los lÃ­mites del intervalo de confianza
    limite_inferior = media_muestral - valor_critico * error_estandar
    limite_superior = media_muestral + valor_critico * error_estandar

    return limite_inferior, limite_superior
```

```python
# Ejercicio 1
nivel_confianza_1 = 0.95
media_muestral_1 = 2.3
desviacion_poblacional_1 = 0.5
tamano_muestra_1 = 50

limite_inferior_1, limite_superior_1 = calcular_intervalo_confianza(media_muestral_1, desviacion_poblacional_1, tamano_muestra_1, nivel_confianza_1)
print(f"Ejercicio 1: Intervalo de confianza del 95%: ({limite_inferior_1}, {limite_superior_1})")
```
