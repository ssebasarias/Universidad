# Algunos elementos a considerar:

1. Es un problema de prueba de hipótesis para la media desconocida, ya que se tiene s que es la desviación estándar muestral.
2. Si el tamaño de la muestra es menor a 30, entonces utilizamos la distribución t
3. Obtener el estadístico de prueba
4. Vamos a obtener el p_Valor
5. Si el p_valor <= alfa, entonces se rechara la hipótesis nula. en caso contrario no existe evidencia estadística que nos permita rechazar Ho

```python
from scipy import stats

# Datos proporcionados
media_muestral = 14
desviacion_estandar_muestral = 4.32
media_hipotesis = 12
tamaño_muestra = 25
alpha = 0.05

# Cálculo del estadístico de prueba (t)
estadistico_prueba = (media_muestral - media_hipotesis) / (desviacion_estandar_muestral / (tamaño_muestra ** 0.5))

# Grados de libertad
grados_libertad = tamaño_muestra - 1

# Cálculo del valor-p usando la distribución t
valor_p = 1 - stats.t.cdf(estadistico_prueba, df=grados_libertad)

# Conclusión basada en el valor de alpha
conclusion = "se rechaza H0" if valor_p <= alpha else "No Se rechaza H0"

estadistico_prueba, valor_p, conclusion
```

```python
from scipy import stats

# Datos proporcionados para el nuevo problema
media_muestral_nueva = 17
desviacion_estandar_muestral_nueva = 4.5
media_hipotesis_nueva = 18
tamaño_muestra_nueva = 48
alpha_nueva = 0.05

# Cálculo del estadístico de prueba (t) para la nueva muestra
estadistico_prueba_nueva = (media_muestral_nueva - media_hipotesis_nueva) / (desviacion_estandar_muestral_nueva / (tamaño_muestra_nueva ** 0.5))

# Grados de libertad para la nueva muestra
grados_libertad_nueva = tamaño_muestra_nueva - 1

# Cálculo del valor-p usando la distribución t para una prueba de dos colas
valor_p_nueva = 2 * (1 - stats.t.cdf(abs(estadistico_prueba_nueva), df=grados_libertad_nueva))

# Conclusión basada en el valor de alpha para la prueba de dos colas
conclusion_nueva = "No se rechaza H0" if valor_p_nueva > alpha_nueva else "Se rechaza H0"

estadistico_prueba_nueva, valor_p_nueva, conclusion_nueva
```

```python
import numpy as np
from scipy.stats import t

# Nivel de significancia
alpha = 0.01

# Tamaño de la muestra
n = 36

# Hipótesis nula
mu_0 = 45

# Datos de las muestras
muestras_datos = [
    {'x_bar': 44, 's': 5.2},
    {'x_bar': 43, 's': 4.6},
    {'x_bar': 46, 's': 5.0}
]

# Realizamos las pruebas de hipótesis para cada conjunto de datos
resultados = []

for datos in muestras_datos:
    # Calculamos el estadístico t
    t_stat = (datos['x_bar'] - mu_0) / (datos['s'] / np.sqrt(n))
    # Calculamos el valor-p para una prueba de cola izquierda
    p_value = t.cdf(t_stat, df=n-1)
    # Dado que la hipótesis alternativa es mu <= 45, usamos la prueba de cola izquierda
    # La hipótesis nula se rechaza si el valor-p es menor que el nivel de significancia
    conclusion = 'Rechazar H0' if p_value < alpha else 'No rechazar H0'
    resultados.append((t_stat, p_value, conclusion))

resultados
```
