```python
import numpy as np
import matplotlib.pyplot as plt

suma= []

dado_1 = np.random.randint(1,7, size = 10)
dado_2 = np.random.randint(1,7, size = 10)

for i, w in enumerate(dado_1):
    suma.append(dado_1[i] + dado_2[i])

print(suma)

k=0
for k in range (2,13):
    print(suma.count(k)/len(suma))
    k=k+1

plt.hist(suma, bins = 20 , alpha = 0.5, label = 'Histograma')

plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generamos una muestra de 1000 lanzamientos de dos dados
suma = []
dado_1 = np.random.randint(1, 7, size=1000)
dado_2 = np.random.randint(1, 7, size=1000)

for i in range(len(dado_1)):
    suma.append(dado_1[i] + dado_2[i])

# Probamos diferentes valores de bins
bins_values = [5, 10, 20, 40, 60]

# Graficamos el histograma para cada valor de bins
for bins in bins_values:
    plt.hist(suma, bins=bins, alpha=0.5, label=f'Bins = {bins}')
    plt.title(f'Histograma con bins = {bins}')
    plt.xlabel('Suma de los dados')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generamos una muestra de 1000 lanzamientos de dos dados
suma = []
dado_1 = np.random.randint(1, 7, size=1000)
dado_2 = np.random.randint(1, 7, size=1000)

for i in range(len(dado_1)):
    suma.append(dado_1[i] + dado_2[i])

# Probamos diferentes valores de alpha
alpha_values = [0.1, 0.3, 0.5, 0.7, 1.0]

# Graficamos el histograma para cada valor de alpha
for alpha in alpha_values:
    plt.hist(suma, bins=20, alpha=alpha, label=f'Alpha = {alpha}')
    plt.title(f'Histograma con alpha = {alpha}')
    plt.xlabel('Suma de los dados')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
prob = np.random.choice([2,3,3,4,4,4,5,5,5,5,6,6,6,6,7,7,7,7,7,8,8,8,8,9,9,9,9,9,10,10,11,11,12], size=15000)
lista = prob.tolist()

# Calcular frecuencia relativa
for k in range(2, 13):
    print(lista.count(k) / len(lista))

# Generar histograma
plt.hist(lista, bins=20, alpha=0.5, label='Histograma')
plt.legend()
plt.grid()
plt.show()
```
