# Expresiones Regulares en Python


Las expresiones regulares, también conocidas como regex o regexp, son patrones de búsqueda que se utilizan para encontrar ciertos patrones de texto dentro de cadenas de caracteres

# Ejemplos básicos

Primero. Caracteres literales: Son aquellos que coinciden exactamente con ellos mismos. Por ejemplo, si la expresión regular es “python” esta coincide con la cadena “python” dentro de un texto.

Segundo.Metacaracteres. Tienen un significado especial dentro de una expresión regular, en lugar de representarse a sí mismos, tienen funciones específicas para realizar tareas como búsquedas o coincidencia de patrones entro de cadenas de texto. Por ejemplo, con ^Inicio se marca el inicio de una línea.

```python
texto = "El gato corre por la calle sin parar. Los perros ladran en la noche. Los niños juegan en el parque."
patron = r"gato"
resultados = re.findall(patron, texto)
print(resultados)
```

```python
palabra1 = 'ta'
palabra2 = 'tazas'
palabra3 = 'pez'

if re.match(palabra1, palabra2): # BUSCAR COICIDENTES.
   print('palabra1 y palabra2 coinciden')
else:
   print('palabra1 y palabra2 no coinciden')

if re.match(palabra1, palabra3):
   print('palabra1 y palabra3 coinciden')
else:
   print('palabra1 y palabra3 no coinciden')
```

```python
import re
cadena = "El pErro saltó sobre la acera."
patron = r"\b[A-Z]\w*"  #
coincidencias = re.findall(patron, cadena)
print(coincidencias) # ['El']
```

```python
import re

cadena = "Hola, ¿cómo estás?"
patron = "cómo"
reemplazo = "estás"
nueva_cadena = re.sub(patron, reemplazo, cadena)
print(nueva_cadena) # "Hola, ¿estás estás?"
```

```python
contactos = """Nombre: Daniel ALejandro
Correo electrónico: daniel.lopez@gmail.com
Teléfono: 316-319-1550

Nombre: Ana López
Correo electrónico: ana_lopez@hotmail.com
Teléfono: 320-987-6543"""
```

```python
import re

# Definimos el patrón de la información de contacto
patron_contacto = r"Nombre: (\w+ \w+)\nCorreo electrónico: (\S+)\nTeléfono: (\d{3}-\d{3}-\d{4})"

# Buscamos todas las coincidencias del patrón en la cadena de texto
coincidencias = re.findall(patron_contacto, contactos)

# Creamos una lista de diccionarios con la información de contacto
contactos_dict = []
for contacto in coincidencias:
    contacto_dict = {
        "nombre": contacto[0],
        "correo_electronico": contacto[1],
        "telefono": contacto[2]
    }
    contactos_dict.append(contacto_dict)

# Imprimimos la lista de diccionarios con la información de contacto
print(contactos_dict)
```
