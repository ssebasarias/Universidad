```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# Configurar Chrome
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

# Inicializar el driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Ir a la página principal
driver.get("https://pricespy.co.uk/")

try:
    # Esperar y aceptar cookies si aparece el botón con data-test
    try:
        accept_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-test="CookieBannerAcceptButton"]'))
        )
        accept_button.click()
        print("✅ Cookies aceptadas correctamente.")
        time.sleep(1)
    except:
        print("⚠️ No apareció el nuevo botón de cookies.")

    # Esperar tarjetas de producto
    product_cards = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[data-testid='CardClickableArea']"))
    )
    print(f"🛒 Se encontraron {len(product_cards)} tarjetas de productos.")

    # Click en el primer producto
    first_product = product_cards[0]
    driver.execute_script("arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});", first_product)
    time.sleep(1)
    first_product.click()
    print("🖱️ Producto clickeado correctamente.")

    # Esperar página de detalles
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "h1"))
    )
    print("✅ Página de detalles del producto cargada.")

except Exception as e:
    print(f"❌ Ocurrió un error: {e}")

# Esperar un poco antes de cerrar
time.sleep(5)
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.common.action_chains import ActionChains


# Configurar Chrome
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

# Inicializar el driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Ir a la página principal
driver.get("https://pricespy.co.uk/")

try:
    # Esperar y aceptar cookies si aparece el botón con data-test
    try:
        accept_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-test="CookieBannerAcceptButton"]'))
        )
        accept_button.click()
        print("✅ Cookies aceptadas correctamente.")
        time.sleep(1)
    except:
        print("⚠️ No apareció el nuevo botón de cookies.")

    # Esperar tarjetas de producto
    product_cards = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[data-testid='CardClickableArea']"))
    )
    print(f"🛒 Se encontraron {len(product_cards)} tarjetas de productos.")

    # Click en el primer producto
    first_product = product_cards[0]
    driver.execute_script("arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});", first_product)
    time.sleep(1)
    first_product.click()
    print("🖱️ Producto clickeado correctamente.")

    # Esperar página de detalles
    WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "h1"))
    )
    print("✅ Página de detalles del producto cargada.")

except Exception as e:
    print(f"❌ Ocurrió un error: {e}")


# Esperar más tiempo para carga total
time.sleep(1)

wait = WebDriverWait(driver, 1)

# Intentar encontrar y hacer clic en el botón "Show more prices"
try:
    print("⏳ Buscando botón 'Show more prices'...")
    
    show_more_button = wait.until(EC.presence_of_element_located(
        (By.XPATH, "//button[./span[contains(text(), 'Show more prices')]]")
    ))

    # Scroll al botón
    print("📜 Haciendo scroll hacia el botón...")
    driver.execute_script("arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});", show_more_button)
    time.sleep(1)

    # Esperar que sea clickeable
    wait.until(EC.element_to_be_clickable((By.XPATH, "//button[./span[contains(text(), 'Show more prices')]]")))

    print("✅ Haciendo clic en el botón...")
    show_more_button.click()

    # Esperar a que cargue el contenido nuevo
    time.sleep(1)

except Exception as e:
    print("❌ No se encontró o no se pudo hacer clic en el botón 'Show more prices'.")
    print("⚠️ Detalles:", e)


# Obtener todos los elementos de precios
precio_elements = driver.find_elements(By.XPATH, '//h4[contains(@class, "StyledPriceLabel")]')

precios = []
actions = ActionChains(driver)

for idx, precio_elem in enumerate(precio_elements):
    try:
        # Hacer scroll hasta el elemento
        actions.move_to_element(precio_elem).perform()
        time.sleep(0.5)  # Pequeña pausa para asegurar que cargue
        precio_text = precio_elem.text.strip()
        precios.append(precio_text)
        print(f"🏪 Tienda #{idx + 1} - Precio: {precio_text}")
    except Exception as e:
        print(f"⚠️ No se pudo extraer el precio para la tienda #{idx + 1}: {e}")

# Mostrar resumen
print(f"\n📊 Total de tiendas encontradas: {len(precios)}")
print(f"💲 Lista de precios: {precios}")

# Esperar un poco antes de cerrar
time.sleep(1)
driver.quit()
```

```python
# Configurar Chrome
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

# Inicializar el driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Ir a la página principal
driver.get("https://pricespy.co.uk/")

try:
    # Esperar y aceptar cookies si aparece el botón con data-test
    try:
        accept_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-test="CookieBannerAcceptButton"]'))
        )
        accept_button.click()
        print("✅ Cookies aceptadas correctamente.")
        time.sleep(1)
    except:
        print("⚠️ No apareció el nuevo botón de cookies.")

    # Esperar tarjetas de producto
    product_cards = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[data-testid='CardClickableArea']"))
    )
    print(f"🛒 Se encontraron {len(product_cards)} tarjetas de productos.")

    # Iterar sobre los primeros 3 productos
    for i in range(min(3, len(product_cards))):
        print(f"\n--- Procesando producto {i + 1} ---")
        card = product_cards[i]

        # Hacer scroll hasta la tarjeta del producto
        driver.execute_script("arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});", card)
        time.sleep(1)

        # Click en la tarjeta del producto
        card.click()
        print("🖱️ Tarjeta de producto clickeada.")

        product_name = None
        try:
            # Esperar a que la página de detalles del producto cargue y obtener el nombre
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "h1.Text--q06h0j.iYVnhg.h2text.StyledDesktopTitle-sc-0-1.lbdHzz"))
            )
            print("✅ Página de detalles del producto cargada.")
            try:
                # Esperar a que el nombre del producto sea visible
                product_name_element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, "h1.Text--q06h0j.iYVnhg.h2text.StyledDesktopTitle-sc-0-1.lbdHzz"))
                )
                product_name = product_name_element.text
                print(f"🏷️ Nombre del producto: {product_name}")
            except Exception as e:
                print(f"⚠️ Error al obtener el nombre del producto: {e}")


        except Exception as e:
            print(f"⚠️ Error al cargar la página de detalles o al obtener el nombre: {e}")
            driver.back()  # Intentar regresar incluso si hay un error
            time.sleep(2)
            continue  # Saltar al siguiente producto si falla la carga de detalles

        if product_name:
            wait = WebDriverWait(driver, 1)

            # Intentar encontrar y hacer clic en el botón "Show more prices"
            try:
                print("⏳ Buscando botón 'Show more prices'...")

                show_more_button = wait.until(EC.presence_of_element_located(
                    (By.XPATH, "//button[./span[contains(text(), 'Show more prices')]]")
                ))

                # Scroll al botón
                print("📜 Haciendo scroll hacia el botón...")
                driver.execute_script("arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});", show_more_button)
                time.sleep(1)

                # Esperar que sea clickeable
                wait.until(EC.element_to_be_clickable((By.XPATH, "//button[./span[contains(text(), 'Show more prices')]]")))

                print("✅ Haciendo clic en el botón...")
                show_more_button.click()

                # Esperar a que cargue el contenido nuevo
                time.sleep(1)

            except Exception as e:
                print("❌ No se encontró o no se pudo hacer clic en el botón 'Show more prices'.")
                print("⚠️ Detalles:", e)

            # Obtener todos los elementos de precios
            precio_elements = driver.find_elements(By.XPATH, '//h4[contains(@class, "StyledPriceLabel")]')

            precios = []
            actions = ActionChains(driver)

            for idx, precio_elem in enumerate(precio_elements):
                try:
                    # Hacer scroll hasta el elemento
                    actions.move_to_element(precio_elem).perform()
                    time.sleep(0.5)  # Pequeña pausa para asegurar que cargue
                    precio_text = precio_elem.text.strip()
                    precios.append(precio_text)
                    print(f"🏪 Tienda #{idx + 1} - Precio: {precio_text}")
                except Exception as e:
                    print(f"⚠️ No se pudo extraer el precio para la tienda #{idx + 1}: {e}")

            # Mostrar resumen del producto actual
            print(f"\n📊 Resumen del producto: {product_name}")
            print(f"💰 Total de tiendas encontradas: {len(precios)}")
            print(f"💲 Lista de precios: {precios}")

        # Regresar a la página principal para el siguiente producto
        driver.back()
        print("🔙 Regresando a la página principal para el siguiente producto.")
        time.sleep(2)

        # Esperar a que las tarjetas de producto vuelvan a ser visibles antes de la siguiente iteración
        if i < 2 and len(product_cards) > i + 1:
            WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[data-testid='CardClickableArea']"))
            )
            # Volver a encontrar las tarjetas (puede que el DOM se haya recargado)
            product_cards = driver.find_elements(By.CSS_SELECTOR, "a[data-testid='CardClickableArea']")

except Exception as e:
    print(f"❌ Ocurrió un error general: {e}")
finally:
    # Cerrar el navegador al finalizar
    driver.quit()
    print("🏁 Proceso completado.")
```
